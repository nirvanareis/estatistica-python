## **Transformação e manipulação de dados com o Pandas**

**Projeto desenvolvido no curso "Pandas: transformação e manipulação de dados" que demonstra comandos de transformação de dados, diferenciação de tipos de informações, métodos de manipulação de valores numéricos, textuais e de datas para conversões de dados gerais.**

```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_hospedagem.json')
dados
```

```python
dados = pd.json_normalize(dados['info_moveis'])   #normalização dos dados
dados
```

```python
colunas = list(dados.columns)  #extração os nomes das colunas
colunas
```

```python
dados = dados.explode(colunas[3:])  #ajuste dos dados em que estão em formato de lista nas colunas de descrição em diante
dados
```

```python
dados.reset_index(inplace = True, drop=True)  #ajuste dos índices do df
dados.head()
```

```python
dados.info()  #verificação os tipos de dados
```

```python
import numpy as np
```

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64) #conversão dos dados de string para inteiro
```

```python
dados.info()  #check
```

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)  #conversão de colunas
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64) #conversão para float
```

```python
dados.info()
```

```python
dados['preco'].apply(lambda x: x.replace('$','').replace(',','').strip()) #remoção os caracteres $ e , da coluna preço
```

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados['preco'] = dados['preco'].astype(np.float64)  #conversão para float
```

```python
dados.info()   #check
```

```python
dados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip())  #conversão de mais de uma coluna
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].astype(np.float64)
```

```python
dados.info()  #check
```

```python
dados['descricao_local'].str.lower()    #manipulação de strings para padronização de letras minúsculas
```

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados['descricao_local']=dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']',' ' ,regex = True)
#remoção todos simbolos exceto os listados
```

```python
dados['descricao_local']=dados['descricao_local'].str.replace('(?<!/w)-(?!/w)',' ' ,regex = True)
dados.head()
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()  #Tokenização de strings
dados.head()
```

```python
dados['comodidades']=dados['comodidades'].str.replace('[/{|}/"]','' ,regex = True) #remoção de caracteres
dados.head()
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')  #Tokenização de strings
dados.head()
```

```python
dados['descricao_vizinhanca']  #dados a serem padronizados
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
```

```python
dados['descricao_vizinhanca']=dados['descricao_vizinhanca'].str.replace('^a-zA-Z0-9\'\-', ' ', regex = True)  #remoção de caracteres
dados['descricao_vizinhanca']=dados['descricao_vizinhanca'].str.replace('(\?<!w)-(?!w)', '', regex = True)
dados.head()
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split() #Tokenização de strings
dados.head()
```
