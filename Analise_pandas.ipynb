### Seleções e Frequências

```python
dado = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
dado.head(8)
```

```python
# seleção dos imóveis classificados como 'apartamento'

selecao_ap = dado['Tipo'] == 'Apartamento'
n1 = dado[selecao_ap]
n1
```

```python
n1 = n1.shape[0]
n1
```

```python
# seleção dos imóveis classificados como 'Casa', 'Casa de Condomínio', 'Casa de Vila' ( símbolo | representa "OU" - "OR")

selecao = (dado['Tipo'] == 'Casa') | (dado['Tipo'] == 'Casa de Condomínio') | (dado['Tipo'] == 'Casa de Vila')
n2 = dado[selecao].shape[0]
n2
```

```python
# selecao dos imóveis com área entre 60 e 100 metros quadrados, incluindo os limites

selecao = (dado['Area'] >= 60) & (dado['Area'] <= 100)
n3 = dado[selecao].shape[0]
n3
```

```python
# seleção dos imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00

selecao = (dado['Quartos'] >= 4) & (dado['Valor'] < 2000)
n4 = dado[selecao].shape[0]
n4
```

```python
print('Nº de imóveis classificados com tipo "Apartamento" -> {}'.format(n1))
print('Nº de imóveis classificados com tipo "Casa", "Casa de condomíno" e "Casa de Vila" -> {}'.format(n2))
print('Nº de imóveis classificados com área entre 60 e 100 metros quadrados -> {}'.format(n3))
print('Nº de imóveis classificados com pelo menos 4 quartos e aluguel menor do que R$2.000,00 -> {}'.format(n4))
```

### Removendo Outliers 

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dado.boxplot(['Valor'])
```

```python
dados[dados['Valor'] >= 500000]
```

```python
valor = dados['Valor']
```

```python
Q1 = valor.quantile(.25)  #1 quartil
Q3 = valor.quantile(.75)  #3 quartil
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
selecao = (valor >= limite_inferior ) & (valor <= limite_superior)
dados_new = dados[selecao]
```

```python
dados_new.boxplot(['Valor'])
```

```python
dados.hist(['Valor'])
dados_new.hist(['Valor'])
```

```python
dados.boxplot(['Valor'], by = ['Tipo'])
```

```python
grupo_tipo = dados.groupby('Tipo')['Valor']
```

```python
type(grupo_tipo)
```

```python
dados_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
    eh_tipo = dados['Tipo'] == tipo
    dentro_limite = (dados['Valor'] >= limite_inferior[tipo]) & (dados['Valor'] <= limite_superior[tipo])
    selecao = eh_tipo & dentro_limite
    dados_selecao = dados[selecao]
    dados_new = pd.concat([dados_new, dados_selecao])
```
