# Manipulando os dados

## Criando colunas numéricas

```python
dados = pd.read_csv(url, sep =';')
dados
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados ['Condominio'] #criando nova coluna de soma do condominio com o aluguel
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

## Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' \
                    + dados['Quartos'].astype(str) + ' quarto(s) e ' + \
                    dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()

#foi preciso convertes os dados inteiros para strings para evitar erros de concatenação
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: 'Sim' if x > 0 else 'Não')  #coluna categórica com o uso de função
```

```python
dados.head()
```

```python
dados.to_csv('dados_completo_dev.csv', index = False, sep = ';') #salvando os dados
```
