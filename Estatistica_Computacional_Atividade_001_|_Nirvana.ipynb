Use a planilha de dados compartilhada pelo professor e calcule o IC 90%, 95% e 99%. Compare a largura dos intervalos.

* Suponha que o SLA da companhia seja até 15 minutos de embarque em média.

* O valor 15 está dentro do IC? O que isso sugere?

* Remova 2 valores extremos da coluna tempo_embarque.

* O IC muda muito? Como interpretar?

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
dados = pd.read_excel("/content/dados_aeroporto (1).xlsx")
dados.head()
```

```python
dados.info()
```

```python
dados['tempo_embarque'] = pd.to_numeric(dados['tempo_embarque'])
```

```python
media_amostra = dados['tempo_embarque'].mean()   #média
s_amostra = dados['tempo_embarque'].std()        #desvio
n = len(dados)                                   #tamanho da amostra
gl = n - 1

print(f'Média: {media_amostra:.2f} Desvio padrão: {s_amostra:.2f} Tamanho da amostra: {n}')
```

```python
nc1 = 0.90  #nivel de confiança de 90%

intervalo1 = stats.t.interval(confidence = nc1,                                     #nível de confiança
                              df = gl,                                              #grau de liberdade
                              loc = media_amostra,                                  #media da amostra
                              scale = stats.sem(dados['tempo_embarque']))           #erro padrão da média (s/sqtr(n))

print(f"Média: {media_amostra:.2f}")
print(f'Intervalo de confiança de 90%: [{intervalo1[0]:.2f}, {intervalo1[1]:.2f}]')
```

```python
nc1 = 0.95  #nivel de confiança de 95%

intervalo2 = stats.t.interval(confidence = nc1,                                     #nível de confiança
                              df = gl,                                              #grau de liberdade
                              loc = media_amostra,                                  #media da amostra
                              scale = stats.sem(dados['tempo_embarque']))           #erro padrão da média (s/sqtr(n))

print(f"Média: {media_amostra:.2f}")
print(f'Intervalo de confiança de 95%: [{intervalo2[0]:.2f}, {intervalo2[1]:.2f}]')
```

```python
nc1 = 0.99  #nivel de confiança de 99%

intervalo3 = stats.t.interval(confidence = nc1,                                     #nível de confiança
                              df = gl,                                              #grau de liberdade
                              loc = media_amostra,                                  #media da amostra
                              scale = stats.sem(dados['tempo_embarque']))           #erro padrão da média (s/sqtr(n))

print(f"Média: {media_amostra:.2f}")
print(f'Intervalo de confiança de 99%: [{intervalo3[0]:.2f}, {intervalo3[1]:.2f}]')
```

É possível observar que o valor 15 está dentro de todos intervalos de confiança testados (90%, 95%, 99%) o que sugere que o valor 15 é um valor plausível para uma verdadeira média da população.

Também é possível notar que existe uma tendência de aumento do intervalo dos dados a medida que aumentamos o nível de confiança do teste. Isso revela que se eu reduzo a precisão do meu teste eu aumento o intervalo para garantir que o valor procurado está ali dentro, no caso contrário ao aumentar o nível de confiança, eu encontro mais segurança, aceitando um intervalo mais largo e uma estimativa menos precisa.

```python
dados_ordenados = dados.sort_values(by = "tempo_embarque")

dados_sem_extremos = dados_ordenados.iloc[1:-1]

dados_sem_extremos.info()
```

```python
media_amostra2 = dados_sem_extremos['tempo_embarque'].mean()   #média
s_amostra2 = dados_sem_extremos['tempo_embarque'].std()        #desvio
n2 = len(dados_sem_extremos)                                   #tamanho da amostra
g2 = n2 - 1

print(f'Média: {media_amostra2:.2f} Desvio padrão: {s_amostra2:.2f} Tamanho da amostra: {n2}')
```

```python
nc1 = 0.90  #nivel de confiança de 90%

intervalo4 = stats.t.interval(confidence = nc1,                                     #nível de confiança
                              df = g2,                                              #grau de liberdade
                              loc = media_amostra2,                                  #media da amostra
                              scale = stats.sem(dados_sem_extremos['tempo_embarque']))           #erro padrão da média (s/sqtr(n))

print(f"Média: {media_amostra2:.2f}")
print(f'Intervalo de confiança de 90%: [{intervalo4[0]:.2f}, {intervalo4[1]:.2f}]')
```

```python
nc2 = 0.95  #nivel de confiança de 95%

intervalo5 = stats.t.interval(confidence = nc2,                                                  #nível de confiança
                              df = g2,                                                           #grau de liberdade
                              loc = media_amostra2,                                              #media da amostra
                              scale = stats.sem(dados_sem_extremos['tempo_embarque']))           #erro padrão da média (s/sqtr(n))

print(f"Média: {media_amostra2:.2f}")
print(f'Intervalo de confiança de 95%: [{intervalo5[0]:.2f}, {intervalo5[1]:.2f}]')
```

```python
nc3 = 0.99  #nivel de confiança de 99%

intervalo6 = stats.t.interval(confidence = nc3,                                                  #nível de confiança
                              df = g2,                                                           #grau de liberdade
                              loc = media_amostra2,                                              #media da amostra
                              scale = stats.sem(dados_sem_extremos['tempo_embarque']))           #erro padrão da média (s/sqtr(n))

print(f"Média: {media_amostra2:.2f}")
print(f'Intervalo de confiança de 99%: [{intervalo6[0]:.2f}, {intervalo6[1]:.2f}]')
```
