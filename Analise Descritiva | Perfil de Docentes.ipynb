
2.   Como a presen√ßa de docentes com necessidades especiais varia de acordo com a combina√ß√£o entre sexo e cor/ra√ßa?
*   Certos grupos, como homens brancos ou mulheres pretas&pardas, podem representar uma parcela mais expressiva do total de docentes com defici√™ncia.





```python
docentes_TP_SEXO_TP_COR_RACA = dados_df[['TP_SEXO','TP_COR_RACA','IN_NECESSIDADE_ESPECIAL']].groupby(['TP_SEXO','TP_COR_RACA']).sum().reset_index()
docentes_TP_SEXO_TP_COR_RACA['grupo'] = (docentes_TP_SEXO_TP_COR_RACA['TP_SEXO'] + " - " + docentes_TP_SEXO_TP_COR_RACA['TP_COR_RACA'])
total_docentes = docentes_TP_SEXO_TP_COR_RACA['IN_NECESSIDADE_ESPECIAL'].sum()

graf_pizza = px.pie(
    docentes_TP_SEXO_TP_COR_RACA,
    names='grupo',
    values='IN_NECESSIDADE_ESPECIAL',
    title=f'Docentes com Necessidades Especiais por Sexo e Cor/Ra√ßa (Total: {total_docentes})'
)
graf_pizza.show()


```


3. Quais s√£o os tipos de defici√™ncia mais frequentes entre os docentes?
*   √â poss√≠vel identificar que a defici√™ncia f√≠sica √© o tipo de necessidade mais frequente dentre o grupo analisado.








```python
ListaColunas=['IN_BAIXA_VISAO','IN_CEGUEIRA','IN_DEF_AUDITIVA','IN_DEF_FISICA','IN_DEF_INTELECTUAL','IN_SURDEZ','IN_SURDOCEGUEIRA','IN_DEF_MULTIPLA','IN_AUTISMO','IN_SUPERDOTACAO']
dados_df['total_deficiencias'] = dados_df[ListaColunas].sum(axis=1)
total_docentes_deficientes = dados_df['IN_NECESSIDADE_ESPECIAL'].sum()
rotulos= {'IN_BAIXA_VISAO': 'Baixa Vis√£o','IN_CEGUEIRA': 'Cegueira','IN_DEF_AUDITIVA': 'Defici√™ncia Auditiva','IN_DEF_FISICA': 'Defici√™ncia F√≠sica',
    'IN_DEF_INTELECTUAL': 'Defici√™ncia Intelectual','IN_SURDEZ': 'Surdez','IN_SURDOCEGUEIRA': 'Surdocegueira','IN_DEF_MULTIPLA': 'Defici√™ncia M√∫ltipla',
    'IN_AUTISMO': 'Autismo','IN_SUPERDOTACAO': 'Superdota√ß√£o'}
docentes_com_deficiencia = dados_df[dados_df['total_deficiencias'] > 0]
apenas_uma = docentes_com_deficiencia[docentes_com_deficiencia['total_deficiencias'] == 1].shape[0]
mais_de_uma = docentes_com_deficiencia[docentes_com_deficiencia['total_deficiencias'] > 1].shape[0]
contagem_deficiencias = dados_df[ListaColunas].sum()

df_plot = pd.DataFrame({'Tipo de Defici√™ncia': contagem_deficiencias.index.map(rotulos),'Quantidade': contagem_deficiencias.values})
graf_linha = px.line(df_plot,x='Tipo de Defici√™ncia',y='Quantidade',markers=True,title='Distribui√ß√£o de Defici√™ncias entre Docentes com Alguma Defici√™ncia')
graf_linha.update_layout(xaxis_tickangle=-45)
graf_linha.show()
print(f"Docentes com apenas uma defici√™ncia: {apenas_uma}")
print(f"Docentes com duas ou mais defici√™ncias: {mais_de_uma}")
```
